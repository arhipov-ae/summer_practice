```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
nemo = cv.imread('./images/nemo0.jpg')
```

```python
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv.cvtColor(nemo, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(nemo)
plt.show()
```

```python
hsv_nemo = cv.cvtColor(nemo, cv.COLOR_RGB2HSV)
plt.imshow(hsv_nemo)
plt.show()
```

```python
light_orange = (1, 190, 200)
dark_orange = (18, 255, 255)
mask = cv.inRange(hsv_nemo, light_orange, dark_orange)
result = cv.bitwise_and(nemo, nemo, mask=mask)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="plasma")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
light_white = (0, 0, 200)
dark_white = (145, 60, 255)

mask_white = cv.inRange(hsv_nemo, light_white, dark_white)
result_white = cv.bitwise_and(nemo, nemo, mask=mask_white)

plt.subplot(1, 2, 1)
plt.imshow(mask_white, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result_white)
plt.show()
```

```python
final_mask = mask + mask_white 
final_result = cv.bitwise_and(nemo, nemo, mask=final_mask)

plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.show()
```
